```python

import numpy as np
import os
import traceback

# Directory containing the binary files
source_directory = "D:/P2M/view_of_delft_PUBLIC/radar/training/velodyne/"
# Directory to save the new .npy files
destination_directory = "D:/vod_npy/vod_radar_test_npy/"

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# Iterate through all the binary files
for i in range(0, 9931):  # Assuming your files are labeled from 00001.bin to 09333.bin
    filename = "{:05d}.bin".format(i)  # Format the filename
    
    try:
        # Check if the file exists
        source_file = os.path.join(source_directory, filename)
        if os.path.exists(source_file):
            # Load the binary file
            scan = np.fromfile(source_file, dtype=np.float32).reshape(-1, 7)
            
            # Define the new filename with .npy extension
            new_filename = os.path.splitext(filename)[0] + '.npy'
            
            # Save the numpy array to file with .npy extension in the destination directory
            np.save(os.path.join(destination_directory, new_filename), scan)
            print(f"File '{filename}' converted and saved successfully.")
        else:
            print(f"File '{filename}' not found.")
    except Exception as e:
        print(f"Error processing file '{filename}':")
        print(traceback.format_exc())

print("Conversion process completed.")
```
